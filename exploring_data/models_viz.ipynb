# MODELS VISUALIZATION

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sector_list = ['Communication services', 'Consumer discretionary',
              'Consumer staples', 'Energy', 'Financials', 
              'Healthcare', 'Industrials', 'Information technology',
              'Utilities']
days_list = [3, 5, 7, 14, 21]
feature_selection = None # f_classif, mutual
```

```python
fig_list = []
for sector_ in sector_list:
    for day_ in days_list:
        if (feature_selection == 'f_classif') or (feature_selection == 'mutual'):
            path = '../models_%s/%s/%i' % (feature_selection, sector_, day_)
        else:
            path = '../models/%s/%i' % (sector_, day_)
        files = os.listdir(os.path.relpath(path))
        for file in files:
            if file.startswith('report'):
                excel_path = os.path.join(path, file)
                df = pd.read_excel(excel_path, sheet_name='report')
                df = df.loc['precision'].drop(['micro avg', 'macro avg', 'weighted avg'])
                plt.figure(figsize=(15,5))
                df.plot(kind='bar', title=file.split('.')[0] + ' %s %i' % (sector_, day_))
```

We can see over all sectors models change through days of prediction. For communication services we can find the strongest precision for S. Bull in 21 days with 60 %. For consummer discretionary we can find over a 50 % in 3 days and better scores in precisions for bull. For the sector healtcare we can find for 7 days an 85 % in W. Bull. The scores for days 5 and 14 in Industrials are 70 % approximetly. In general without any feature extraction bull's classification has better scores than bear's classification. On the other hand, when we make an ANOVA the scores of bear improve in general although these scores are very low, but as a remarkable, for 21 days in healthcare we achieve a 90 % of precision for S. Bull.
